*   ***Here we have Increased Brightness and Contrast of the videos.***
*   ***Increasing only brightness didn't have much effect on thhe videos, it just whitens the videos, but nothing else.***
*   ***On the other hand, increasing contrast gives colors to the brightened videos, thereby, making them seeable even if shot in the dark.***
*   ***Every folder has to be input one by one, therefore the code has to be run 11 times***







```python
import os
import cv2

def brighten_and_contrast_videos(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    for root, _, files in os.walk(input_folder):
        for file in files:
            input_path = os.path.join(root, file)

            output_subfolder = os.path.relpath(root, input_folder)
            output_subfolder_path = os.path.join(output_folder, output_subfolder)
            os.makedirs(output_subfolder_path, exist_ok=True)

            output_path = os.path.join(output_subfolder_path, file)

            video = cv2.VideoCapture(input_path)

            width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
            height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
            fps = video.get(cv2.CAP_PROP_FPS)

            fourcc = cv2.VideoWriter_fourcc(*'mp4v')
            out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

            while True:
                ret, frame = video.read()

                if not ret:
                    break

                frame = cv2.convertScaleAbs(frame, alpha=7, beta=40)

                out.write(frame)

            video.release()
            out.release()

            print(f"Processed: {input_path} -> {output_path}")

# Example usage
input_folder = "/content/drive/MyDrive/Colab Notebooks/ARID_v1_5_211015.zip (Unzipped Files)/clips_v1.5/Wave"
output_folder = "/content/drive/MyDrive/Colab Notebooks/Test/Wave"
brighten_and_contrast_videos(input_folder, output_folder)
```
